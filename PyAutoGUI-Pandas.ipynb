```python
# https://pyautogui.readthedocs.io/en/latest/quickstart.html - comandos do pyautogui

# bibliotecas para automatizar o mouse, o teclado e o monitor

import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 0.5 # tempo entre os comandos


# Abrir o sistema da empresa 

pyautogui.hotkey("ctrl", "t") # abrir aba
pyperclip.copy("https://drive.google.com/drive/u/1/folders/1CW6U4b8ZBfifJl4ynyGLouqRGx80HEL4") # copiar link
pyautogui.hotkey("ctrl", "v") # colar link
pyautogui.press("enter") # teclando o enter para entrar no sistema

time.sleep(5) # carregar o sistema até que ele abra


# Baixar a base de dados

pyautogui.click(x=600, y=529) # click no arquivo
pyautogui.click(x=600, y=529, button='right')
pyautogui.click(x=842, y=1234) # click em download

time.sleep(5) # esperar o download



# Levar a base para o python

import pandas as pd

tabela_de_dados = pd.read_excel(r"C:\Users\Windows 10 Pro\Downloads\Vendas - Dez.xlsx") # armazenar os dados na variável
display(tabela_de_dados) # exibir os dados




# Calcular o faturamento e quantidade de produtos

faturamento = tabela_de_dados["Valor Final"].sum() # soma dos valores finais
quantidade_produtos = tabela_de_dados["Quantidade"].sum() # soma das quantidades


# Abrir email

pyautogui.hotkey("ctrl", "t") # abrir aba
pyperclip.copy("https://mail.google.com/mail/u/2/#inbox") # copiar link
pyautogui.hotkey("ctrl", "v") # colar link
pyautogui.press("enter") # teclando o enter para entrar no sistema
time.sleep(2)
# Enviar email

pyautogui.click(x=167, y=315) # click para criar um novo email
time.sleep(1)
pyautogui.click(x=3166, y=1275)
pyperclip.copy("e.kaique@aluno.ifsp.edu.br") # copiar texto com caracteres especiais
pyautogui.hotkey("ctrl", "v") # colar o texto
pyautogui.press("tab") # muda o campo - assunto
pyperclip.copy("Relatório de Vendas") # copiar texto com caracteres especiais
pyautogui.hotkey("ctrl", "v") # colar o texto
pyautogui.press("tab") # muda o campo - texto

texto = f'''
Prezados, bom dia

O faturamento de ontem foi de: R$ {faturamento:,.2f}
A quantidade de produtos foi de: {quantidade_produtos:,}

Abs
Kaique Santos''' # armazenar a mensagem na variável 

pyperclip.copy(texto) # copiar o texto da variável
pyautogui.hotkey("ctrl", "v") # colar o texto
pyautogui.hotkey("ctrl", "enter") # enviar o email

```
